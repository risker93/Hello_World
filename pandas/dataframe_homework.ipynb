잘 열린다

```python
#readline() 
f = open("file_test.txt", "r")
while True:
    line = f.readline()
    if not line:
        break
   
    item = line.split()
    i = int(item[0])
    name = item[1]
    age = int(item[2])
    adress = item[3]
    ph_nu = item[4]
    dataframe.loc[i]=[name, age, adress, ph_nu]
    
        
    
f.close()
```
