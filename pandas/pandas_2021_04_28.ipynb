```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
df= titanic.loc[:,['age','sex','class','fare','survived']]
df
#보통 앞에 원인이 되는 요소들을 넣고 맨뒤에 결과를 갖는 요소를 넣음
```

```python
grouped = df.groupby(['class'])
```

```python
grouped
```

```python
grouped.age
```

```python
age_mean= grouped.age.mean()
age_mean
```

```python
age_std =grouped.age.std()
age_std
```

```python
for k, g in grouped.age:
    group_zscore = (g - age_mean.loc[k]) / age_std.loc[k]
    #z 스코어 공식을 보고 위의 한줄 코드를 짤수 있느냐 없느냐가 중요
    print(k)
    print(group_zscore)
```

```python
def z_score(x):
    return (x -x.mean())/x.std()
```

```python
grouped.age.transform(z_score)
```

```python
trans_age = grouped.age.transform(z_score)
trans_age.loc[[1,10,2]]
```

```python
trans_age.loc[1:9]
```
