#3. 미분, 순간 변화율과 기울기

- 딥러닝을 이해하는 데 가장 중요한 수학 원리는 미분이라고 할 수 있음

- 너무 미세해서 실제로 움직이는 게 아니라 방향만 드러내는 정도의 순간적인 변화만 있을 것임

- 이 순간의 변화를 놓고 순간 변화율이라는 이름을 붙임

- 순간 변화율은 어느 쪽을 향하는 방향성을 지니고 있으므로, 이 방향을 따라 직선을 길게 그려주면 그래프와 맞닿는 접선이 그려짐

- 이 선이 바로 이 점에서의 기울기가 됨

접선의 기울기가 되는 x축의 값을 arg min 이라고 한다

- 미분을 한다는 것은 쉽게 말해 이 '순간 변화율' 을 구한다는 것이

- 미분계수: 어느 순간에 어떤 변화가 일어나고 있는지를 숫자로 나타낸 것

- 이 미분 계수는 곧 그래프에서의 기울기를 의미함

캡처.JPG


- 이 그래프에서 x 값의 증가량은 b -a

- y값의 증가량은 f(b) - f(a)

직선 AB의 기울기 = y값의 증가량 / x값의 증가량

f(b) - f(a) / b - a

f(a + dx) - f(a) / dx

- 이때 직선 AB의 기울기를 A와 B 사이의 '평균 변화율' 이라고도 부름

- 미분을 배우고 있는 우리에게 필요한 것은 순간 변화율임

- 순간 변화율: x의 증가량 dx 가 0에 가까울 만큼 아주 작을때의 순간적인 기울기를 말함

lim dx->0  f(a+dx)-f(a)/dx

- lim dx->0  : x의 증가량이 0에 가까울 만큼 작을때 라는 뜻

- 기울기는 y의증가량/x의 증가량  
이므로 순간 기울기는  lim dx->0  y값의 증가량/ x값의 증가량  
이것은 lim dx->0 dy/dx   = dy/dx 라고도 쓸수 있음 

- 함수 f(x)를 미분하라는 것은 f'(x) 또는 d/dx  f(x) 로 표기함

캡처2.JPG


- 딥러닝을 공부하는 과정에서 자주 만나게 되는 중요한 4가지 미분의 성질

1. f(x) = a 에서 a 가 상수일때 미분 값은 0임
2. f(x) = x 일 때의 미분 값은 1임
3. f(x) = ax 에서 a 가 상수이면 미분값은 a임 
4. f(x) = x^a 에서 a가 자연수이면 미분값은 ax^a-1 이됨


#4. 편미분

- 독립변수 : 'x값이 변함에 따라 y값도 변한다' 는 이 정의 안에서, 독립적으로 변할 수 있는 x 값

- 다중 선형회귀(multiple linear regression): x값이 여러 개 필요할 때

2. 가장 훌륭한 예측선이란?

- 선형 회귀를 공부하는 과정은 이 점들의 특징(feature)을 가장 잘 나타내는 선을 그리는 과정과 일치함.

- 여기에서 선은 직선이므로 곧 일차 함수 그래프임 y=ax+b

#4. 코딩으로 확인하는 최소 제곱

```python
import numpy as np
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
```

- 이제 최소 제곱근 공식으로 기울기 a와 y절편의 b 값을 구해보자

- x의 모든 원소의 평균을 구하는 넘파이 함수는 mean()임

- mx 라는 변수에 x원소들의 평균값을, my에 y원소들의 평균값을 입력

```python
mx = np.mean(x)
my = np.mean(y)
```

- 최소 제곱근 공식 중 분모의 값, 즉 'x의 각 원소와 x의 평균 값을의 차를 제곱하라' 는 파이썬 명령을 만들 차례

- 다음과 같이 divisor 라는 변수를 만들어 구현할 수 있음

```python
divisor = sum([(i-mx)**2 for i in x])
```

- 이제 분자에 해당하는 부분을 구함

- x와 y의 편차를 곱해서 합한 값을 구하면 됨

- 다음과 같이 새로운 함수를 정의하여 dividend 변수에 분자의 값을 저장함

```python
def top (x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx)*(y[i] - my)
  return d
dividend = top(x, mx, y, my)
```

- 임의의 변수 d의 초깃값을 0으로 설정한 뒤 x의 개수만큼 실행함

- d에 x의 각 원소와 평균의 차, y의 각 원소와 평균의 차를 곱해서 차례로 더하는 최소제곱법을 그대로 구현함.


- 이제 위에서 구한 분모와 분자를 계산하여 기울기 a 를 구함

a = dividend/ divisor

- a를 구하고 나면 y 절편을 구하는 공식을 이용해 b를 구할 수 있음

b = my- (mx*a)
